```python
import pandas as pd

# Assuming 'dataset' contains your data with columns: movie_id, user_id, ratings

# Group by 'movie_id' and count the number of ratings for each movie
movie_ratings_count = data.groupby('movie_id')['rating'].count()

# Sort the movies based on the number of ratings in descending order
sorted_movies = movie_ratings_count.sort_values(ascending=False)

# Display the movies sorted by the number of ratings
sorted_movies
```

```python
top_n = 40  # Change this number to select a different number of movies

top_movies = sorted_movies.head(top_n)  # Select the top N movies by rating count
top_movies_ids = top_movies.index.tolist()  # Get the movie IDs of the top N movies

# Slice the dataset to include only the top N movies
filtered_data = data[data['movie_id'].isin(top_movies_ids)]

# Display the filtered dataset containing only the top N movies
filtered_data
```

```python
from sklearn.metrics.pairwise import cosine_similarity

# Calculating cosine similarity
user_similarity = cosine_similarity(movie_feat, dense_output=False)
```

```python
user_similarity
```

```python
# user_similarity=user_similarity.toarray()
```

```python
movie_feat.head()
```

```python
def predict_ratings(user_id, movie_id):
    user_row = movie_feat.iloc[:,movie_id]
    sim_scores = user_similarity[user_id]
    # movie_ratings = movie_feat.loc['movie_id']==movie_id]

    # Calculate the weighted sum of ratings
    prediction = user_row.dot(sim_scores) / sim_scores.sum()
    return prediction

# Example: Predicting rating for user_id=10, movie_id=10
predicted_rating = predict_ratings(100, 10)

print(f"Predicted rating for user  on movie 10: {predicted_rating}")
```

```python
new_li = []
for pos_i,i in enumerate(movie_feat.index):
  for pos_j,j in enumerate(movie_feat.columns):
    new_li.append([i,j,predict_ratings(pos_i,pos_j)])
new_li
```

```python
from sklearn.metrics import mean_squared_error
```

```python
predicted_df = pd.DataFrame(new_li, columns=['user_id', 'movie_id', 'predicted_rating'])

# Merge original dataset with predicted ratings
merged_df = pd.merge(data, predicted_df, on=['user_id', 'movie_id'], how='inner')

print(len(merged_df))

```

```python
# Calculate MSE
mse = mean_squared_error(merged_df['rating'], merged_df['predicted_rating'], squared=False)
print(f"Root Mean Squared Error (RMSE): {mse**0.5}")
```
