```python
# Assuming the dataset is in a CSV file named 'ratings.csv'
file_path = '/content/drive/My Drive/final_netflix_price.csv'
data = pd.read_csv(file_path)

# Display the first few rows to understand the data
print(data.head())
```

```python
import pandas as pd

# Assuming 'dataset' contains your data with columns: movie_id, user_id, ratings

# Group by 'movie_id' and count the number of ratings for each movie
movie_ratings_count = data.groupby('movie_id')['rating'].count()

# Sort the movies based on the number of ratings in descending order
sorted_movies = movie_ratings_count.sort_values(ascending=False)

# Display the movies sorted by the number of ratings
print(sorted_movies)
```

```python
top_n = 20  # Change this number to select a different number of movies

top_movies = sorted_movies.head(top_n)  # Select the top N movies by rating count
top_movies_ids = top_movies.index.tolist()  # Get the movie IDs of the top N movies

# Slice the dataset to include only the top N movies
filtered_data = data[data['movie_id'].isin(top_movies_ids)]

# Display the filtered dataset containing only the top N movies
print(filtered_data)
```

```python
from scipy.sparse import csr_matrix

movie_features_df_matrix = csr_matrix(movie_feat.values)
```

```python
from scipy.sparse import csr_matrix

movie_features_df_matrix = csr_matrix(movie_features_df.T.values)

from sklearn.neighbors import NearestNeighbors


model_knn = NearestNeighbors(metric = 'cosine', algorithm = 'brute')
model_knn.fit(movie_features_df_matrix)
```

```python
query_index = np.random.choice(movie_features_df.T.shape[0])
print(query_index)
distances, indices = model_knn.kneighbors(movie_features_df.T.iloc[query_index,:].values.reshape(1, -1), n_neighbors = 6)
```

```python
for i in range(0, len(distances.flatten())):
    if i == 0:
        print('Recommendations for {0}:\n'.format(movie_features_df.T.index[query_index]))
    else:
        print('{0}: {1}, with distance of {2}:'.format(i, movie_features_df.T.index[indices.flatten()[i]], distances.flatten()[i]))
```

```python
# Assuming data is your dataset containing columns user_id, rating, movie_id, title

for i in range(0, len(distances.flatten())):
    if i == 0:
        movie_id = movie_features_df.T.index[query_index]
        title = data[data['movie_id'] == movie_id]['title'].values[0]
        print('Recommendations for {0}:\n'.format(title))
    else:
        movie_id = movie_features_df.T.index[indices.flatten()[i]]
        title = data[data['movie_id'] == movie_id]['title'].values[0]
        print('{0}: {1}, with distance of {2}:'.format(i, title, distances.flatten()[i]))
```
